```python
import requests
from scrapy.http import TextResponse 
import pandas as pd
```

```python
url = "https://www.menu.am/"
```

```python
page = requests.get(url)
response = TextResponse(url=page.url,body=page.text,encoding="utf-8")
```

```python
full_urls = [url[:-1]+i for i in response.css("div[class='fl list-logo'] > a::attr(href)").extract()]
```

```python
print(full_urls)
```

```python
name = []
price = []
for i in full_urls[:2]:##scraping only for two pages
    page = requests.get(i)
    response = TextResponse(url=page.url,body=page.text,encoding="utf-8")
    name.extend(response.css("div[class='fl prod-content']>a::text").re("\s(.*)"))
    price.extend(response.css("div[class='price']::text").re("\d+"))
```

```python
df=pd.DataFrame({'name':name,'price':price})##making dataframe
```
